#### e. Understand the trends/seasonality of sales, quantity by category, location, month, week, day, time, channel, payment method etc…

```python
ord_item_orders['month'] = ord_item_orders['order_purchase_timestamp'].dt.strftime('%b')
```

```python
ord_item_orders_prod = pd.merge(left = ord_item_orders, right = products,on='product_id',how = 'left')
ord_item_orders_prod.head(2)
```

```python
#product category-wise sales

ord_item_orders_prod.groupby(by='product_category_name')[['Quantity','total_cost']].sum().reset_index()
```

```python
#month-wise sales

ord_item_orders_prod.groupby(by='month')[['Quantity','total_cost']].sum().reset_index()
```

```python
#week-wise sales

ord_item_orders_prod['Week_Num'] = ord_item_orders_prod['order_purchase_timestamp'].dt.isocalendar().week
ord_item_orders_prod.head()
```

```python
#weekly sales trend
ord_item_orders_prod.groupby('Week_Num')[['Quantity','total_cost']].sum().reset_index().rename(columns = {'total_cost':'Sales'})
```

```python
geo_loc
```

```python
#location-wise sales

ord_item_orders_cust = pd.merge(left=ord_item_orders,right =cust,how ='left',on='customer_id' )
ord_item_orders_cust.head(2)
```

```python
#state-wise sales

result = ord_item_orders_cust.groupby(by='customer_state')[['Quantity','total_cost']].sum().reset_index()
result_sorted = result.sort_values(by=['Quantity', 'total_cost'], ascending=False)
result_sorted
```

```python
#city-wise sales

result =ord_item_orders_cust.groupby(by='customer_city')[['Quantity','total_cost']].sum().reset_index()
result_sort = result.sort_values(by=['Quantity','total_cost'],ascending=False)
result_sort
```

```python
#channel-wise sales

ord_item_orders_pay = pd.merge(left = ord_item_orders,right =order_pay,how= 'left',on='order_id' )

ord_item_orders_pay.head()
```

```python
result = ord_item_orders_pay.groupby(by='payment_type')[['Quantity','total_cost']].sum().reset_index()
result_sorted =  result.sort_values(by=['Quantity','total_cost'],ascending=False)
result_sorted
```

```python
plt.figure(figsize=(10, 6))  
sns.barplot(data=result, x='payment_type', y='total_cost', palette="viridis")

plt.xticks(rotation=45)

# Adding labels and title
plt.xlabel('Payment Type')
plt.ylabel('Total Cost')
plt.title('Total Cost by Payment Type')
plt.show()
```
